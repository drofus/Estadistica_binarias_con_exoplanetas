```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("binary_stars.xlsx")
df2 = pd.read_csv("confirmed_planets.csv")
df3 = pd.read_csv("Exo_planets_DATA_EU.csv")
```

```python
lista = []
talis = []

for i in range(len(df2)):
    for j in range(len(df)):
        if df2["hostname"][i] == df['Name Nasa'][j]:
            lista.append(i)

for i in range(len(df3)):
    for j in range(len(df)):
        if df3['star_name'][i] == df['NAme EU'][j]:
            talis.append(i)
```

```python
prueba = df2.iloc[lista]
aprueb = df3.iloc[talis]
Tabla_NASA_binarias = prueba.reset_index()
Tabla_EU_binarias = aprueb.reset_index()
```

# USANDO LA BASE DE DATOS DE NASA ARCHIVE

```python
Tabla_NASA_NO_binarias = df2.drop(lista)

Tabla_NASA_binarias = Tabla_NASA_binarias.drop_duplicates(subset=['hostname'])
Tabla_NASA_NO_binarias = Tabla_NASA_NO_binarias.drop_duplicates(subset=['hostname'])

df2 = df2.drop_duplicates(subset=['hostname'])

Tabla_NASA_binarias = Tabla_NASA_binarias.reset_index()
Tabla_NASA_NO_binarias = Tabla_NASA_NO_binarias.reset_index()
df2 = df2.reset_index()
```

1. st_age -> edad de la estrellas (Gyr)
2. st_logg -> gravedad superficial
3. st_lum -> luminosidad estelar ($log_{10}(L_\odot)$)
4. st_teff -> temperatura superficial (K)
5. st_mass -> masa estelar ($M_\odot$)
6. st_met -> metalicidad (dex)
7. st_rad -> radio estelar ($R_\odot$)
8. st_vsin -> velocidad de rotación (km/s)

```python
#edades
edad = df2['st_age']
n_edad = Tabla_NASA_NO_binarias['st_age']
b_edad = Tabla_NASA_binarias['st_age']

#gravedad superficial
logg = df2['st_logg']
n_logg = Tabla_NASA_NO_binarias['st_logg']
b_logg = Tabla_NASA_binarias['st_logg']

#luminosidad estelar
lum = df2['st_lum']
n_lum = Tabla_NASA_NO_binarias['st_lum']
b_lum = Tabla_NASA_binarias['st_lum']

#temperatura superficial
teff = df2['st_teff']
n_teff = Tabla_NASA_NO_binarias['st_teff']
b_teff = Tabla_NASA_binarias['st_teff']

#masa estelar
mass = df2['st_mass']
n_mass = Tabla_NASA_NO_binarias['st_mass']
b_mass = Tabla_NASA_binarias['st_mass']

#metalicidad
met = df2['st_met']
n_met = Tabla_NASA_NO_binarias['st_met']
b_met = Tabla_NASA_binarias['st_met']

#radio estelar
rad = df2['st_rad']
n_rad = Tabla_NASA_NO_binarias['st_rad']
b_rad = Tabla_NASA_binarias['st_rad']

#velocidad de rotación
vsin = df2['st_vsin']
n_vsin = Tabla_NASA_NO_binarias['st_vsin']
b_vsin = Tabla_NASA_binarias['st_vsin']
```

## Graficas con edad vs ....

```python
eliminar = []
b_eliminar = []
n_eliminar = []

for i in range(len(edad)):
    if np.isnan(edad[i]) == True or np.isnan(logg[i]) == True:
        eliminar.append(i)

for i in range(len(b_edad)):
    if np.isnan(b_edad[i]) == True or np.isnan(b_logg[i]) == True:
        b_eliminar.append(i)

for i in range(len(n_edad)):
    if np.isnan(n_edad[i]) == True or np.isnan(n_logg[i]) == True:
        n_eliminar.append(i)



edad_a = edad.drop(eliminar)
logg_a = logg.drop(eliminar)

b_edad_a = b_edad.drop(b_eliminar)
b_logg_a = b_logg.drop(b_eliminar)

n_edad_a = n_edad.drop(n_eliminar)
n_logg_a = n_logg.drop(n_eliminar)
```

### Edad vs logg

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(edad_a, logg_a, '*')
#ax1.set_yscale("log")
ax1.set_xscale("log")
ax2.plot(n_edad_a, n_logg_a, '*')
ax2.plot(b_edad_a, b_logg_a, '*')
#ax2.set_yscale("log")
ax2.set_xscale("log")
```

# BOXPLOTS

```python
def get_box_plot_data(labels, bp):
    rows_list = []

    for i in range(len(labels)):
        dict1 = {}
        dict1['label'] = labels[i]
        dict1['lower_whisker'] = bp['whiskers'][i*2].get_ydata()[1]
        dict1['lower_quartile'] = bp['boxes'][i].get_ydata()[1]
        dict1['median'] = bp['medians'][i].get_ydata()[1]
        dict1['upper_quartile'] = bp['boxes'][i].get_ydata()[2]
        dict1['upper_whisker'] = bp['whiskers'][(i*2)+1].get_ydata()[1]
        rows_list.append(dict1)

    return pd.DataFrame(rows_list)
```
